**Review Problem 2**

**Concept:** *reading CSV data, matplotlib, line graph*

**Task:**

Write a function `read_and_plot_column(filename, column_name)` that:
* Opens a CSV file (do **not** use pandas)
* Extracts the values from the specified column
* Casts all values to `float`
* Plots the values using `plt.plot()`
* Adds a title
* Displays the plot

````python
# Your Solution
def read_and_plot_column(filename, column_name):
    '''  
    A function read_and_plot_column() reads a csv andplots the values using plt.plot()
    '''
    with open(filename, "r") as f:
        # Header extraction, but next(reader) with csv module is a cleaner solution
        header = f.readline().strip().split(",")

        # Initialize dictionary with empty lists
        data = {key: [] for key in header}

        # Process rows
        for line in f:
            values = line.strip().split(",")
            # key and value are both lists with alligning values
            # So we can zip them together
            for key, value in zip(header, values):
                 try:
                     # We are adding to a list!
                     # In my code the value is always string at the beginning
                     data[key].append(int(value))
                 except ValueError:
                     try:
                         data[key].append(float(value))
                     except ValueError:
                         data[key].append(value)
                         
        dictionary = {k: np.array(v) for k, v in data.items()}

        if column_name in dictionary.keys():
            col = [float(x) for x in dictionary[column_name]]

            plt.plot(col)
            plt.title("Problem 2: line graph")
            plt.xlabel("Day Number")
            plt.ylabel(column_name)
            plt.show()
````

**Task:**

Using `df_weather`:
* Write a helper function to extract time (in hours) from a datetime string and convert it to decimal hours (float)
* Convert time using:

    `hours + (minutes / 60) + (seconds / 3600)`
* Create a new DataFrame with the same index as `df_weather`
* Add the following columns:
  * `sunrise` (converted to float)
  * `sunset` (converted to float)
  * `tempmin` (copied directly from `df_weather`)
* Plot all columns together using `.plot()`


**Sample Output:**

This is similar to how your new dataframe should look before plotting:

```
   start_time  end_time  price
P1       8.25     17.50     20
P2       8.50     17.75     35
P3       9.00     18.00     15
P4       9.25     18.25     60
P5       9.50     18.50     50
```

````python
# Your Solution

def hour_extract(date_col):
    '''  
    A function hour_extract() that extarct time (in hours) from a datetime string
    and convet it to decimal hours (float).
    
    '''
    date_col = pd.to_datetime(date_col)

    return (
        date_col.dt.hour +
        date_col.dt.minute / 60 +
        date_col.dt.second / 3600
    )

df_new = pd.DataFrame(index=df_weather.index)

df_new["sunrise"] = hour_extract(df_weather["sunrise"])
df_new["sunset"] = hour_extract(df_weather["sunset"])
df_new["tempmin"] = df_weather["tempmin"]

fig, ax = plt.subplots()
ax.plot(df_new)
ax.set_title("Problem 10: line plot")
ax.set_ylabel("Temperature in Celcius degrees")
ax.set_xlabel("Date")
ax.legend(df_new.columns) # legend will match, but in my opinion this method is not error proof
plt.show()
````

**Concept:** *Filtering `DataFrame` with boolean conditions*


**Task:**


Write a function `filter_rows_matching_pattern(df, column, pattern)` that:
* Returns rows where the specified column contains the given pattern
* Handles missing values
* Prints the number of matching rows

**Sample Output:**

If your selected column contained matching text values, the following could be your output:
```
Found 1 records of color containing blue.
Found 1 records of product containing hat.
Found 1 records of color containing red.
```

````python
# Your Solution
def filter_rows_matching_pattern(df, column, pattern):
    '''  
    Function  filter_rows_matching_pattern() returns rows where the specified column contains the given pattern.  

    '''

# .str = vectorized string operations. It only works when values are strings.
    mask = df[column].str.contains(pattern, na=False) # If the value is missing (NaN), treat it as False

    filtered_df = df[mask]

    print(f"Found {len(filtered_df)} records of {column} containing {pattern}.")
    
    return filtered_df
````

````python
# Your Solution
def create_df_categorical(df_source, column_name):
    ''' 
    Function create_df_categorical creates a new dataframe with counts of each category and percentage of each category
    '''

    counts = df_source[column_name].value_counts()
    counts_norm = df_source[column_name].value_counts(normalize = True)  

    # Pandas matched rows by indexlabel. That's the reason I allgined value_counts() togetehr.
    return pd.DataFrame({"count" : counts, "pct": counts_norm })

````

**Problem 17**

**Concept:** *Filtering records by category.*


**Task:**

Write a function `filter_weather(df, conditions, temp_min, temp_max)` that:
* Filters rows where:
  * conditions contain the given string
  * `tempmax` ≤ `temp_max`
  * `tempmin` ≥ `temp_min`
* Prints intermediate counts
* Returns the filtered DataFrame

**Sample Output:**

If your data matched the filtering conditions step by step, the following could be your output:
```
4 records with conditions Blue
3 records with conditions Blue and tempmax <= 50
2 records with conditions Blue and 20 <= temp <= 50
Found 2 records with conditions of Blue and temp between 20 and 50.
```

````python
# Your Solution
def filter_weather(df, conditions, temp_min, temp_max):
    ''' 
    A function filter_weather() filters rows, prints intermediate counts, and returns the filtered DataFrame.
    
    '''
    mask1 = df["conditions"].str.contains(conditions, na=False)
    mask2 = df["tempmax"] <= temp_max
    mask3 = df["tempmin"] >= temp_min
    bool_mask = mask1 & mask2 & mask3

    filtered_df_condition = df[mask1]
    filtered_df_tempmax = df[mask1 & mask2]  
    filtered_df_fin = df[bool_mask] 

    print(f"{len(filtered_df_condition)} records with conditions {conditions}")
    print("")
    print(f"{len(filtered_df_tempmax)} records with conditions {conditions} and tempmax <= {temp_max}")
    print("")
    print(f"{len(filtered_df_fin)} records with conditions of {conditions} and {temp_min} <= temp <= {temp_max}")
    print("")
    print(f"Found {len(filtered_df_fin)} records with {conditions} and temp between {temp_min} and {temp_max}.")
    print("")

    return filtered_df_fin
````

Write a function `create_standardized_values(df_source, columns, row_index=None)` that:

**Sample Output:**

      high_value  low_value

If you call the function

you should get the following as the output:
```
      high_value  low_value
2020         1.0       1.00
2021         1.2       1.10
2022         1.4       1.20
2023         1.6       1.40
2024         1.8       1.50

```

````python
# Your Solution
def create_standardized_values(df_source, columns, row_index = None):
    '''  
    A function create_standardized_values relative to given column  and returns 

    '''
    base_row = df_source.loc[row_index] if row_index is not None else df_source.iloc[0]

    new_df = {col: df_source[col] / base_row[col] for col in columns}

    # I could also return df_source[columns].div(base_row[columns])
    return pd.DataFrame(new_df)        
````
